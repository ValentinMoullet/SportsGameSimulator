```python
import pandas as pd
```

```python
GAME_INFO_FILE = 'data/football-events/ginf.csv'
OUTPUT_TRAINING_IDS = 'data/football-events/training_ids.csv'
OUTPUT_TEST_IDS = 'data/football-events/test_ids.csv'
TRAINING_SET_RATIO = 0.7
ALL_LEAGUES = ['D1', 'E0', 'F1', 'I1', 'SP1']
```

```python
game_info_df = pd.read_csv(GAME_INFO_FILE)
game_info_df.head()
```

```python
all_training_ids = []
all_test_ids = []
for league in ALL_LEAGUES:
    league_df = game_info_df[game_info_df['league'] == league].sort_values(by='date')
    all_ids = league_df['id_odsp'].values
    nb_games_training = int(len(all_ids) * TRAINING_SET_RATIO)
    nb_games_test = len(all_ids) - nb_games_training
    
    training_ids = all_ids[:nb_games_training]
    test_ids = all_ids[nb_games_training:]
    
    all_training_ids.extend(training_ids)
    all_test_ids.extend(test_ids)
    
print("nb_games_training:", len(all_training_ids))
print("nb_games_test:", len(all_test_ids))
```

```python
training_df = pd.DataFrame(all_training_ids)
training_df.columns = ['training_id']
training_df.to_csv(OUTPUT_TRAINING_IDS)

test_df = pd.DataFrame(all_test_ids)
test_df.columns = ['test_id']
test_df.to_csv(OUTPUT_TEST_IDS)
```
